# Data Cleaning

```python
# Removing unnecessary columns
# remove 'id'
df.drop('id', axis=1, inplace=True)
```

```python
# inspecting the missing values
# price - row check,
df_null = df[df['price'].isnull()]
```

```python
for i in df.columns:
    print("___",i,"___")
    print(df_null[i].value_counts())
    print("\n")
```

```python
# We observe that null values are from "taxi" by df['name'] only.
df.isnull().sum()
```

```python
# Fill in the missing values
# we fill the price field having NaN with 0
df_pr = df.copy()
df_pr['price'].fillna(0.0, inplace=True)
```

```python
# Remove the missing values
# we remove the price field having NaN
df_rm = df.dropna()
```

```python
df_pr.info()
```

```python
df_rm.info()
```

```python
# continue with df_pr i.e., where we fill NAN with 0.0
# introducing a column called Booked having values 1 for Yes and 0 for No
book = []
for i in df_pr['price']:
    if i>0:
        book.append(1)
    else:
        book.append(0)

df_pr['booked'] = pd.Series(np.array(book))
df_pr
```

Renaming the columns

```python
# name -> ride_option
df_pr.rename(columns = {'name': 'ride_option'}, inplace = True)
```

# Data transformation

```python
# Convert the 'time_stamp' column to datetime format
time = pd.to_datetime(df_pr['time_stamp']/1000, unit='s')
```

```python
time
```

```python
# Transform the 'time_stamp' column and break it down to month, weekday, hour
# fill 'month', 'day', 'hour' into the dataset
df_pr['Month'] = time.dt.month
df_pr['Day'] = time.dt.dayofweek
df_pr['Hour'] = time.dt.hour
df_pr.head()
```

```python
# Remove 'time_stamp'
df_pr.drop('time_stamp',axis=1,inplace=True)
df_pr.head()
```

At this point, it will be better to make a copy of the dataset.

```python
# Create dummy of df_pr
df_pr_dummy = df_pr.copy(deep=True)
```

```python
# One hot encode (for categorical features)
olist = ["cab_type", "destination", "source", "product_id","ride_option"]
olistn = []
for i in range(df_pr_dummy.shape[1]):
    for j in olist:
        if df_pr_dummy.columns[i] == j:
            olistn.append(i)
```

```python
olistn
```

```python
# OneHotEncoding the nominal string values
```

```python
df_pr_dummy = pd.get_dummies(df_pr_dummy, columns=olist, dtype='uint8')
```

# Modeling and Training

```python
# dividing the dataset based on predictor and target variables
X = df_pr_dummy.drop('price',axis=1)
y = df_pr_dummy['price']
```

```python
# importing necessary libraries from Sci-Kit learn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.svm import LinearSVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold

from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error as rm, r2_score, mean_absolute_error
```

```python
# splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# # Scaling the dataset
# scaler = StandardScaler()
# scalery = StandardScaler()
# # we will scale after the train_test_split
# scaler.fit(X_train)
```

```python
# # scaler.fit(x_train)
# # scalery.fit(y_train)
# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)
```

```python
# Ridge Regression
```

```python
ridge = Ridge(alpha=1, solver="cholesky")
ridge.fit(X_train,y_train)
```

```python
# training score
y_pred_ridge = ridge.predict(X_test)
ridge.score(X_train,y_train)
```

```python
# test metrics
print("MAE: ",mean_absolute_error(y_test, y_pred_ridge))
print("MSE: ",rm(y_test, y_pred_ridge))
print("RMSE: ",np.sqrt(rm(y_test, y_pred_ridge)))
print("R2 Score: ",r2_score(y_test, y_pred_ridge))
```

```python
# Linear Regression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
# training score
y_pred_lr = lr.predict(X_test)
lr.score(X_train,y_train)
```

```python
# test metrics
print("MAE: ",mean_absolute_error(y_test, y_pred_lr))
print("MSE: ",rm(y_test, y_pred_lr))
print("RMSE: ",np.sqrt(rm(y_test, y_pred_lr)))
print("R2 Score: ",r2_score(y_test, y_pred_lr))
```

```python
# Support Vector Machine
```

```python
svr = LinearSVR(epsilon=1.5,loss="squared_epsilon_insensitive", max_iter=70000, dual=False)
svr.fit(X_train,y_train)
```

```python
# training score
y_pred_svr = svr.predict(X_test)
svr.score(X_train,y_train)
```

```python
# test metrics
print("MAE: ",mean_absolute_error(y_test, y_pred_svr))
print("MSE: ",rm(y_test, y_pred_svr))
print("RMSE: ",np.sqrt(rm(y_test, y_pred_svr)))
print("R2 Score: ",r2_score(y_test, y_pred_svr))
```

```python
# Random Forest Regressor
```

```python
rf = RandomForestRegressor(n_estimators = 10, random_state = 0)  #tried doing for 1000 trees, but takes time(so kept it as 10)
rf.fit(X_train, y_train)
```

```python
# training score
y_pred_rf = rf.predict(X_test)
rf.score(X_train,y_train)
```

```python
# test metrics
print("MAE: ",mean_absolute_error(y_test, y_pred_rf))
print("MSE: ",rm(y_test, y_pred_rf))
print("RMSE: ",np.sqrt(rm(y_test, y_pred_rf)))
print("R2 Score: ",r2_score(y_test, y_pred_rf))
```
